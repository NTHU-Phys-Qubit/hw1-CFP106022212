# 1D Random walk

We will concider two kinds of 1D random talk.
* $dx$ is uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2.

For both kinds of the 1D random walk, you should do the following:
* Plot a single trajectory.
* Plot the histogram of the end point for an ensemble of trajectories.

```python
b = 0
c = np.zeros(1000)
c[0] = b
for i in range(1,1000):
    b += np.random.random()-0.5
    c[i]=b
plt.plot(c)
```

```python
people = 1000

a = np.zeros(people)
for i in range(1,100):
    a += np.random.random(people)-0.5
    
plt.hist(a,bins=50)
```

```python
b = 0
c = np.zeros(1000)
c[0] = b
for i in range(1,1000):
    b += np.random.randint(2)*2-1
    c[i]=b
plt.plot(c)
```

```python
people = 1000

a = np.zeros(people)
for i in range(1,100):
    a += np.random.randint(2 ,size = people)*2-1
    
plt.hist(a)#,bins=50)
```
